```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import Math
```

```python
baris = np.array([2,3])
kolom = np.array([[1],[4]])
```

```python
print(baris)
print(kolom)
```

```python
display(Math(sym.latex(sym.simplify(baris))))
```

```python
display(Math(sym.latex(sym.simplify(kolom))))
```

```python
baris = np.array(baris).flatten()
kolom = np.array(kolom).flatten()
```

```python
plt.plot([0, baris[0]],[0, baris[1]], "r", label= "Vektor Baris")
plt.plot([0, kolom[0]],[0, kolom[1]], "b", label= "Vektor Kolom")
plt.axis("square")
plt.axis([-5,5,-5,5])
plt.grid()
plt.legend()
plt.show()
```
